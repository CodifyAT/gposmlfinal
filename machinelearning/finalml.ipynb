```python
import pandas as pd
df=pd.read_csv("Data1.csv")
df.head(10)
```

```python
del df["Unnamed: 1"]
```

```python
df = df.rename(columns={'Unnamed: 0': 'index'})
```

```python
df
```

```python
from statistics import mean,median,mode
```

```python

plhh=[]
for i in range(68):
    count=0
    plh=[]
    a=[]
    for j in range(len(df["Plant Height(cm)"])):
        if(df["index"][j]==i):
            #print(df.loc[j])
            plh.append(df["Plant Height(cm)"][j])
    c=mean(plh)
    for i in range(len(plh)):
        plhh.append(c)
```

```python
df["Plant Height(cm)"]=plhh
```

```python

subpop=[]
plhh=[]
for i in range(68):
    count=0
    sbp=[]
    plh=[]
    for j in range(len(df["index"])):
        if(df["index"][j]==i):
            #print(df.loc[j])
            sbp.append(df["Subpopulation"][j])
    print(sbp)
    if(len(sbp)==2):sbp[1]=sbp[0]
    else:
        a=len(sbp)
        b=mode(sbp)
        for i in range(a):sbp[i]=b
    for i in sbp:
        subpop.append(i)
    #print(sbp)
```

```python
leg={'A':'1','C':'3','G':'4','T':'4'}
seqq=[]
for i in df['Sequences']:
    seq=""
    for j in i:seq+=leg[j]
    seqq.append(int(seq))
```

```python
df["Sequences"]=seqq
df["Subpopulation"]=subpop
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
```

```python
df1=pd.read_csv("Encodseqdf.csv")
```

```python
x1=df['Sequences'].values
y1=df['Subpopulation'].values
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.20, random_state=42)
x_train1 = x_train1.reshape(-1, 1)
x_test1 = x_test1.reshape(-1, 1)
```

```python
subpoprf = KNeighborsClassifier()  # Corrected class name
subpoprf.fit(x_train1, y_train1)
y_pred1 = subpoprf.predict(x_test1)
y_pred1
```

```python
subpopac=accuracy_score(y_test1,y_pred1)
subpopac
```

```python

x2 = df['Sequences'].values
y2 = df['Plant Height(cm)'].values
x_train2, x_test2, y_train2, y_test2 = train_test_split(x2, y2, test_size=0.25)
x_test2 = x_test2.reshape(-1, 1)
x_train2 = x_train2.reshape(-1, 1)
plhrf = RandomForestRegressor(n_estimators=100,random_state=35,max_depth=10)
plhrf.fit(x_train2, y_train2)
y_pred2 = plhrf.predict(x_test2)
y_pred2
```

```python
heightacc=r2_score(y_test2,y_pred2)
heightacc
```

```python

x3 = df['Sequences'].values
y3 = df['Yield'].values
x_train3, x_test3, y_train3, y_test3 = train_test_split(x2, y2, test_size=0.25)
x_test3 = x_test3.reshape(-1, 1)
x_train3 = x_train3.reshape(-1, 1)
yld = RandomForestRegressor()
yld.fit(x_train3, y_train3)
y_pred3 = yld.predict(x_test3)
ylac=r2_score(y_test3,y_pred3)
ylac
```

```python
a = input("Enter a 9 digit sequence: ")
leg = {'A': '1', 'C': '3', 'G': '4', 'T': '4'}
seqq = ""
for j in a:
    seqq += leg[j]
score = np.array([int(seqq)])  # Convert the sequence to an integer array
score = score.reshape(-1, 1)
prh = plhrf.predict(score)
sbpp = subpoprf.predict(score)
yeils=yld.predict(score)
print("Entered sequence: ", a)
print("Predicted height: ", prh[0])
print("Accuracy for the height: ",heightacc*100,"%")
print("Predicted subpopulation: ", sbpp[0])
print("Accuracy of the subpopulation :",subpopac*100,"%")
print("Predicted Yield:",yeils[0])
print("Accuracy of Yield: ",ylac*100,"%")
```
